```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
n = 10001
k = 30
data1 = np.random.randn(n) -1
data2 = 3*np.random.randn(n) 
data3 = np.random.randn(n) +1

# histograms
y1,x1 = np.histogram(data1,bins=k)
#x is equal to the number of boundaries for the bins
# we need to check the bins' centers
xx1 = (x1[1:]+x1[:-1])/2

y2,x2 = np.histogram(data2,bins=k)
xx2 = (x2[1:]+x2[:-1])/2
y3,x3 = np.histogram(data3,bins=k)
xx3 = (x3[1:]+x3[:-1])/2
# line histogram
plt.plot(xx1,y1,label = 'Data1')

plt.plot(xx2,y2,label = 'Data2')

plt.plot(xx3,y3,label = 'Data3')
plt.title('Histogram with ' +str(k)+ ' bins')
plt.legend()
plt.xlabel('Data values')
plt.ylabel('Data counts')

```

### Mean

```python
Mean1 = np.mean(data1)
Mean2 = np.mean(data2)
Mean3 = np.mean(data3)

Median1 = np.median(data1)
Median2 = np.median(data2)
Median3 = np.median(data3)
```

```python
plt.plot(xx1,y1,'b', xx2,y2,'r', xx3,y3,'k')
plt.plot([Mean1,Mean1],[0,max(y1)],'b--',label = '%s' %round(Mean1,2))
plt.plot([Mean2,Mean2],[0,max(y2)],'r--',label = '%s' %round(Mean2,2))
plt.plot([Mean3,Mean3],[0,max(y3)],'k--',label = '%s' %round(Mean3,2))
plt.plot([Median1,Median1],[0,max(y1)],'r^-',label = '%s' %round(Median1,2))
plt.plot([Median2,Median2],[0,max(y2)],'k^-',label = '%s' %round(Median2,2))
plt.plot([Median3,Median3],[0,max(y3)],'b^-',label = '%s' %round(Median3,2))

#in empirical data we are not expecting the mean to be exactly in the peak
plt.legend()
plt.xlabel('Data values')
plt.ylabel('Data counts')
```
